# Preprocessing the Images

Preprocessing the image to remove unwanted data from it and only contain required image.

```python
import os
from PIL import Image

# Folder containing the images
folder_path = "/content/drive/MyDrive/Data/Data 2"
output_folder_path = "/content/drive/MyDrive/Data/Train Cropped Image"

# Ensure the output folder exists
os.makedirs(output_folder_path, exist_ok=True)

# Predefined coordinates for the yellow box
x1, y1, x2, y2 = 66, 68, 926, 606

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".png") or filename.endswith(".jpg"):  # Add other image formats if needed
        image_path = os.path.join(folder_path, filename)

        # Load the image
        image = Image.open(image_path)

        # Crop the image
        cropped_image = image.crop((x1, y1, x2, y2))

        # Save the cropped image to the output folder
        cropped_image_path = os.path.join(output_folder_path, f"{filename}")
        cropped_image.save(cropped_image_path)

        #print(f"Cropped and saved: {cropped_image_path}")
```
